```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import gradio as gr

# Load your GTD dataset
df = pd.read_csv("globalterrorismdb_0718dist.csv", encoding="ISO-8859-1", low_memory=False)

# Fix invalid months
df['imonth'] = df['imonth'].replace(0, 1)

# Create year_month datetime
df['year_month'] = pd.to_datetime(df['iyear'].astype(str) + '-' + df['imonth'].astype(str), format='%Y-%m')

# Filter out unknown countries if any
df = df[df['country_txt'].notnull()]

# Aggregate attacks by country & month
country_month_df = df.groupby(['country_txt', 'year_month']).size().reset_index(name='num_attacks')

# Create full date range for time series (monthly start)
all_months = pd.date_range(country_month_df['year_month'].min(), country_month_df['year_month'].max(), freq='MS')
countries = country_month_df['country_txt'].unique()

# Build full time series per country filling missing months with 0 attacks
ts_df = pd.DataFrame()

for country in countries:
    temp = country_month_df[country_month_df['country_txt'] == country].set_index('year_month').reindex(all_months, fill_value=0)
    temp = temp.rename(columns={'num_attacks': 'attacks'})
    temp['country_txt'] = country
    temp['year_month'] = temp.index
    ts_df = pd.concat([ts_df, temp.reset_index(drop=True)])

ts_df.reset_index(drop=True, inplace=True)

# Rolling attack features
ts_df['attacks_last_3'] = ts_df.groupby('country_txt')['attacks'].rolling(window=3, min_periods=1).sum().shift(1).reset_index(level=0, drop=True)
ts_df['attacks_last_6'] = ts_df.groupby('country_txt')['attacks'].rolling(window=6, min_periods=1).sum().shift(1).reset_index(level=0, drop=True)

# Target label: will attack happen in next 3 months?
ts_df['future_attacks_3'] = ts_df.groupby('country_txt')['attacks'].rolling(window=3, min_periods=1).sum().shift(-3).reset_index(level=0, drop=True)
ts_df['target'] = (ts_df['future_attacks_3'] > 0).astype(int)

# Drop rows with NaNs in features/target
ts_df = ts_df.dropna(subset=['attacks_last_3', 'attacks_last_6', 'target'])

# Features and target
features = ['attacks_last_3', 'attacks_last_6']
X = ts_df[features]
y = ts_df['target']

# Train/test split stratified by target
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Model training
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

print(f"Test accuracy: {model.score(X_test, y_test):.3f}")

# Build Gradio app to predict by country
countries_sorted = sorted(ts_df['country_txt'].unique())

def predict_attack_by_country(country_name):
    country_data = ts_df[ts_df['country_txt'] == country_name].sort_values('year_month')
    latest = country_data.iloc[-1]
    X_input = latest[features].values.reshape(1, -1)
    prob = model.predict_proba(X_input)[0, 1]
    pred = model.predict(X_input)[0]
    if pred == 1:
        return f"âš ï¸ {country_name} likely to have terrorist attacks in next 3 months! Probability: {prob:.2f}"
    else:
        return f"âœ… {country_name} unlikely to have terrorist attacks in next 3 months. Probability: {prob:.2f}"

app = gr.Interface(
    fn=predict_attack_by_country,
    inputs=gr.Dropdown(choices=countries_sorted, label="Select Country"),
    outputs="text",
    title="Country-level Terror Attack Predictor",
    description="Select a country to predict whether it will experience terrorist attacks in the next 3 months."
)

if __name__ == "__main__":
    app.launch()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the GTD dataset
df = pd.read_csv("globalterrorismdb_0718dist.csv", encoding="ISO-8859-1", low_memory=False)

# Clean up: remove unknown groups and unknown countries for clarity
df = df[(df['gname'] != 'Unknown') & (df['country_txt'].notnull())]

# Aggregate: count number of attacks by each group per country
group_country_attacks = df.groupby(['gname', 'country_txt']).size().reset_index(name='attack_count')

# Optional: show top 10 groups by total attacks for visualization
top_groups = group_country_attacks.groupby('gname')['attack_count'].sum().sort_values(ascending=False).head(10).index

# Filter to only top 10 groups for clean heatmap
filtered_data = group_country_attacks[group_country_attacks['gname'].isin(top_groups)]

# Pivot for heatmap (rows: groups, cols: countries)
pivot_table = filtered_data.pivot(index='gname', columns='country_txt', values='attack_count').fillna(0)

# Display top groups attack counts per country table (text)
print(pivot_table)

# Plot heatmap
plt.figure(figsize=(16, 8))
sns.heatmap(pivot_table, cmap="Reds", linewidths=0.5, linecolor='gray')

plt.title("ðŸ”¥ Terrorist Groups Attack Counts by Country (Top 10 Groups)", fontsize=16)
plt.ylabel("Terrorist Group", fontsize=13)
plt.xlabel("Country", fontsize=13)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Step 1: Aggregate the data per country
country_df = df.groupby('country_txt').agg({
    'eventid': 'count',  # total attacks
    'nkill': 'sum',
    'nwound': 'sum'
}).rename(columns={'eventid': 'total_attacks'}).reset_index()

# Optional: Add casualties column
country_df['casualties'] = country_df['nkill'] + country_df['nwound']

# Step 2: Scale the features
features = country_df[['total_attacks', 'nkill', 'nwound', 'casualties']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 3: KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
country_df['cluster'] = kmeans.fit_predict(scaled_features)

# Step 4: Plot results
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=country_df, x='total_attacks', y='casualties', hue='cluster', palette='tab10')
plt.title('K-Means Clustering of Countries Based on Terrorism Data')
plt.xlabel('Total Attacks')
plt.ylabel('Total Casualties')
plt.grid(True)
plt.show()
```
